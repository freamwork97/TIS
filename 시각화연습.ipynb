```python
plt.plot([1,2,3],[4,1,2], label='line1')
plt.plot([1,2,3],[5,6,0], label='line2')
plt.legend(loc = 'upper right')
```

```python
plt.plot([1,2,3],[4,1,2], color='black',label='line1')
plt.plot([1,2,3],[5,6,0], color='red',label='line2')
plt.legend(loc = 'upper right')
```

```python
plt.plot([1,2,3],[4,1,2], linestyle='dashed',label='line1')
plt.plot([1,2,3],[5,6,0], linestyle='dotted',label='line2')
plt.legend(loc = 'upper right')
```

```python
plt.plot([1,2,3],[4,1,2],marker = '*', markersize='8', label='line1')
plt.plot([1,2,3],[5,6,0],marker = 'D', markersize='8', label='line2')
plt.legend(loc = 'upper right')
```

```python
plt.scatter([1,2,3],[4,1,2], label='scatter1')
plt.scatter([1,2,3],[5,6,0], label='scatter2')
plt.legend(loc='upper right')
```

```python
plt.scatter([1,2,3],[4,1,2], marker='D', s = 80,color='red', label='scatter1')
plt.scatter([1,2,3],[5,6,0], marker='*', s = 80,color='green', label='scatter2')
plt.legend(loc='upper right')
```
