# concat

```python
import pandas as pd
df1 = pd.DataFrame({'A':[1,2,3],
                    'B':[1,2,3]})
df2 = pd.DataFrame({'A':[4,5],
                    'B':[4,5]})

display(df1)
print('-'*10)
display(df2)
```

```python
print('ignore_index = True')
display(pd.concat([df1,df2],axis=0, ignore_index=True))
print('-'*10)
print('ignore_index = False')
display(pd.concat([df1,df2],axis=0, ignore_index=False))
```

```python
df3 = pd.concat([df1,df2],axis=0, ignore_index=False)
df3.loc[1]
```

```python
# 연습
# 두개의 데이터프레임을 위 아래로 연결
# 새로운 통합 인덱스를 부여
import numpy as np

df1 = pd.DataFrame(np.arange(12).reshape(4,3),index=['a','b','c','d'])
df2 = pd.DataFrame(np.arange(12).reshape(4,3),index=['A','B','C','D'])

display(df1)
print('-'*20)
display(df2)

df3=pd.concat([df1,df2],axis=0,ignore_index=True)
print('-'*20)
df3
```

```python
df1 = pd.DataFrame({'key':[1,2,3],'x':[1,2,3]})
df2 = pd.DataFrame({'key':[2,3,4],'y':[2,3,4]})
display(df1)
print('-'*10)
display(df2)
```

```python
display(pd.merge(df1,df2, on='key',how='inner'))
```

```python
display(pd.merge(df1,df2, on='key',how='left'))
```

```python
display(pd.merge(df1,df2, on='key',how='right'))
```

```python
display(pd.merge(df1,df2, on='key',how='outer'))
```
