```python
# 연습문제
# 다음 데이터프레임의 'A' 컬럼과 'B' 컬럼을 뺀 다음
# 다음 그 결과에 절대값을 취한 값들로 이루어진
# 새로운 칼럼 'C'를 생성
# 조건 apply 메서드 활용
# 절대값 np.abs
import numpy as np
df = pd.DataFrame(np.random.rand(12).reshape(6,2), columns=['A', 'B'])
print(df)
```

```python
df['C'] = df['A'] - df['B']
df['C'] = np.abs(df['C'])
print(df)
```

```python
def a(a):
  return np.abs(a)
df['C'] = df['A']-df['B']
df['C']=df['C'].apply(a)
df
```

```python
df['C'] = np.abs(df['A']-df['B'])
df
```

```python
# df 전체 요소대상
df = df+1
df
```

```python
# 람다 함수 적용
# 1~5 인덱스별 최대치와 최소치의 차

df.apply(lambda x:x.max() - x.min(), axis=1)
```
